```python
import matplotlib.pyplot as plt
import skimage.data as skdata
import skimage.feature as skfeat
import skimage.color as skcolor
import skimage.io as skio
```

# Step 1: Perform Canny edge detection on the `skdata.astronaut()` image.

- call skimage.astronaut() to obtain the image
- convert it to grayscale
- perform Canny edge detection (`skfeat.canny()` function) on the grayscale image -- select the parameters to optimize the edge detection to find only the edges of major salient objects in the scene such as the person, helmet, flag, and space shuttle model (you won't be able to do this perfectly, but do the best you can).  
- display the result of the Canny edge detection as a binary image.

```python
I_astronaut = skcolor.rgb2gray(skdata.astronaut())
```

```python
sigma = 2
low_threshold = 0.3
high_threshold = 0.6
mode = 'nearest'

edges = skfeat.canny(I_astronaut, sigma=sigma, low_threshold=low_threshold, high_threshold=high_threshold, mode=mode)
```

```python
plt.figure(figsize=(10, 5))

plt.subplot(1, 2, 1)
plt.imshow(I_astronaut, cmap='gray')
plt.title('Original Image')
plt.axis('off')

plt.subplot(1, 2, 2)
plt.imshow(edges, cmap='gray')
plt.title('Canny Edge Image')
plt.axis('off')

plt.tight_layout()
plt.show()
```

# Step 2: Justify your parameter selection

Explain why you selected the parameters that you did and justify why they are appropriate for the given edge detection task and context.

The parameter `sigma=2` is a moderate value that allows to ignore the minor details and allow the algorithm to focus on the major edges. The `low_threshold=0.3` and `high_threshold=0.6` are selected to allow the algorithm to detect the edges of the major objects in the image while smoothing the details selected by the `sigma` parameter.

We can see that the result allows us to detect the edges of the major objects in the image. I'm not completely happy on the way the lower part of the rocket is detected nor the face of the astronaut, but I think that the result is good enough for the purpose of the exercise.


# Step 3: Repeat for the `mushroom.jpg` image.

- load the image file `mushroom.jpg` provided with this assignment and convert it to grayscale
- perform Canny edge detection on the grayscale image -- select the parameters to find only the edges of the five mushrooms in the foreground that are in focus (you won't be able to do this perfectly, but do the best you can).
- display the result of the Canny edge detection as a binary image.

```python
# Since this image is a little larger, it may display with a fuzzy appearance if you don't increase the 
# resolution of figure rendering a bit.  Before you call plt.imshow(), call plt.figure(dpi=300).  
# This will result in a clearer display of the binary edge image.

I_mushroom = skio.imread('data/mushroom.jpg', as_gray=True)
```

```python
sigma = 2.73
low_threshold = 0.2
high_threshold = 0.48
mode = 'nearest'

edges = skfeat.canny(I_mushroom, sigma=sigma, low_threshold=low_threshold, high_threshold=high_threshold, mode=mode)
```

```python
plt.figure(figsize=(10, 5), dpi=300)

plt.subplot(1, 2, 1)
plt.imshow(I_mushroom, cmap='gray')
plt.title('Original Image')
plt.axis('off')

plt.subplot(1, 2, 2)
plt.imshow(edges, cmap='gray')
plt.title('Canny Edge Image')
plt.axis('off')

plt.tight_layout()
plt.show()
```

# Step 4: Justify your parameter selection

Explain why you selected the parameters that you did and justify why they are appropriate for the given edge detection task and context.

The parameter `sigma=2.73` is a value (higher than in the previous image) that allows to ignore some of the most noticeable mushrooms that are blurred but still present strong edges. The `low_threshold=0.2` and `high_threshold=0.48` are selected to allow the algorithm to detect the edges of the main mushrooms, sadly it still picks two parts of mushrooms that are blurred but have strong edges.
